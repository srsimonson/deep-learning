# Image Augmentation

Image augmentation is a strategy for generating new data points when you don't have enough. For example, take one cat image, you can resize it, flip it, rotate it, etc. You now have 4x different valid cat images from one image.

```python
# %pip uninstall -y opencv-python opencv-python-headless numpy
# %pip install "numpy>=1.26,<2.0"
# %pip install "opencv-python<4.13"
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
img = mpimg.imread('../../data-sets/cat_image.jpg')
```

```python
plt.imshow(img)
```

## Rescaling

```python
print(img.shape)
```

```python
new_height = int(img.shape[1] * 0.1)   # 119
new_width = int(img.shape[0] * 0.1)    # 212

img_rescale = resize(img, (new_height, new_width))

plt.imshow(img_rescale)
```

```python
newer_height = int(img_rescale.shape[1] * 0.5)   # 119
newer_width = int(img_rescale.shape[0] * 0.5)    # 212

img_rescale_again = resize(img_rescale, (newer_height, newer_width))
plt.imshow(img_rescale_again)
```

## Flipping

```python
horizontal_flip = np.fliplr(img)
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img)
plt.imshow(vertical_flip)
```

## Rotation

```python
from skimage import transform
trans_image = transform.rotate(img, random.uniform(-40, 40))
plt.imshow(trans_image)
```

## Adding Noise

```python
img_noise = random_noise(img, mode='s&p', clip = True)
plt.imshow(img_noise)

# Doesn't distort the image, but adds random pixel noise.
```

```python
img_noise2 = random_noise(img_noise, mode='s&p', clip = True)
```

```python
plt.imshow(img_noise2)
```

```python
img_noise3 = random_noise(img_noise2, mode='s&p', clip = True)
plt.imshow(img_noise3)
```
